```python
def lm(X,y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
    X_train = sm.add_constant(X_train)
    X_test = sm.add_constant(X_test)
    ols_model = sm.OLS(y_train, X_train).fit()
    print(ols_model.summary())
    y_pred = ols_model.predict(X_test)

    mse = mean_squared_error(y_test, y_pred)
    return
lm(X=X,y=y)
```

```python
def make_state_plot(state,year,data,statefips):
    data['fips'] = data['fips'].astype(str)
    state_df = data[(data['fips'].str.startswith(statefips))]
    state_df = state_df[state_df['Year'] == year]

    

    values = state_df['Count.B'].tolist()
    fips = state_df['fips'].tolist()
    for i in values:
        if i <= 0:
            print(i)
            values[i] = 0

    endpts = list(np.mgrid[min(values):max(values):8j])

    colorscale = sns.color_palette("mako", 9).as_hex()

    fig = ff.create_choropleth(
        fips=fips,
        values=values,
        scope=[state],
        show_state_data=True,
        colorscale=colorscale,
        binning_endpoints=endpts,
        round_legend_values=True,
        plot_bgcolor='rgb(229,229,229)',
        paper_bgcolor='rgb(229,229,229)',
        legend_title='Grocery Stores per County',
        county_outline={'color': 'rgb(255,255,255)', 'width': 0.5},
        exponent_format=True,
    )
    

    fig.layout.template = None
    fig.show()
    return
make_state_plot('North Carolina',2017,df2,'37')
```
